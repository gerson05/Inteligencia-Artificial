```python
# Exploración de datos de características
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/annotated_data/feature_data.csv')
df.head()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['angle_hip'], kde=True)
plt.title('Distribución de ángulo de cadera')
plt.xlabel('Ángulo de cadera')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='label', y='angle_hip', data=df)
plt.title('Ángulo de cadera por actividad')
plt.xlabel('Actividad')
plt.ylabel('Ángulo de cadera')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='label', data=df)
plt.title('Cantidad de muestras por clase')
plt.xlabel('Actividad')
plt.ylabel('Cantidad')
plt.show()
```

```python
if len(df.select_dtypes(include='number').columns) > 1:
    plt.figure(figsize=(6,5))
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
    plt.title('Matriz de correlación')
    plt.show()
```

## Ejemplo de características extraídas por frame
A continuación se muestra una tabla de ejemplo (primeros 5 frames) y una gráfica de la distribución de los ángulos y distancias extraídos por frame.

```python
# Cargar el archivo de características detalladas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cambia el path si tu archivo tiene otro nombre
features = pd.read_csv('../data/annotated_data/feature_data_detailed.csv')
features.head()
```

```python
# Selecciona solo las columnas de ángulos y distancias
cols = [col for col in features.columns if any(x in col for x in ['angle', 'dist'])]

# Histograma para cada característica
features[cols].hist(bins=20, figsize=(15, 8), layout=(2, int(len(cols)/2)+1))
plt.suptitle('Distribución de ángulos y distancias extraídos por frame')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Boxplot de cada característica por tipo de actividad (si hay columna 'label')
if 'label' in features.columns:
    for col in cols:
        plt.figure(figsize=(8,4))
        sns.boxplot(x='label', y=col, data=features)
        plt.title(f'{col} por actividad')
        plt.xlabel('Actividad')
        plt.ylabel(col)
        plt.show()
```
